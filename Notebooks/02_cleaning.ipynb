### Step 2 : Data Cleaning
##### Input : We will put raw dataset that is dft-road-casualty-statistics-collision-last-5-years.csv
##### Process : Cleaning data , dropping columns and row , handling missing data 
##### Output : Meaningful and proper new cleaned dataset

```python
import numpy as np
import pandas as pd 
df = pd.read_csv('..\\Data\\dft-road-casualty-statistics-collision-last-5-years.csv')
df.info()
df.shape
```

```python
dropping_columns_leakage = ['enhanced_severity_collision','collision_injury_based','collision_adjusted_severity_serious',
                            'collision_adjusted_severity_slight','did_police_officer_attend_scene_of_accident']
dropping_columns_identifier = ['collision_index','collision_ref_no','lsoa_of_accident_location','local_authority_ons_district',
                               'local_authority_highway','local_authority_highway_current','local_authority_district',
                               'first_road_number','second_road_number','police_force']
dropping_columns_historic = ['junction_detail_historic','pedestrian_crossing_human_control_historic',
                             'pedestrian_crossing_physical_facilities_historic','carriageway_hazards_historic']
dropping_columns_highmissing = ['junction_control','trunk_road_flag','location_easting_osgr','location_northing_osgr']
columns_drop = dropping_columns_identifier+dropping_columns_leakage+dropping_columns_historic+dropping_columns_highmissing
df = df.drop(columns = columns_drop)
df.shape
```

```python
df = df.replace(-1,np.nan)
(df == -1).sum()*100/len(df)
```

```python
dropping_missing_rows=['longitude','latitude','speed_limit','light_conditions','weather_conditions','urban_or_rural_area']
df = df.dropna(subset=dropping_missing_rows)
filling_missing_rows = ['junction_detail','second_road_class','pedestrian_crossing','road_surface_conditions',
                        'special_conditions_at_site','carriageway_hazards',]
for row in filling_missing_rows:
    df[row] = df[row].fillna(df[row].mode()[0])
print(df.isnull().sum()*100/len(df))
```

```python
df.to_csv('../Data/cleaned_collision.csv', index=False)
print("collsion is saved or nahh")
print(df.shape)
```

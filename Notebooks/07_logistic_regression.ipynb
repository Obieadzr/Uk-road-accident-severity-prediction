```python
import numpy as np
import pandas as pd 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,ConfusionMatrixDisplay
import matplotlib.pyplot as plt
pd.read_csv('../Data/Scaled_collision.csv') 
X_train=np.load('../Data/X_train.npy')
X_test=np.load('../Data/X_test.npy')
Y_train=np.load('../Data/Y_train.npy')
Y_test=np.load('../Data/Y_test.npy')
```

```python
model_lr = LogisticRegression(max_iter=1000,class_weight='balanced',random_state=42)
model_lr.fit(X_train,Y_train)
```

```python
y_pred = model_lr.predict(X_test) 
print(y_pred[:10])
print(Y_test[:10])
```

```python
print(accuracy_score(y_pred,Y_test))
print(classification_report(Y_test, y_pred, target_names=['Fatal','Serious','Slight']))
```

```python
ConfusionMatrixDisplay.from_predictions(Y_test,y_pred,display_labels=['Fatal', 'Serious', 'Slight'],
    cmap='Blues')
plt.show()
```

#### Logistic Regression Results

- Accuracy: 50% — lower than a naive baseline but expected because 
  class_weight='balanced' forces the model to attempt all three classes.

- Fatal Recall: 0.63 — the model catches 63% of actual fatal accidents. 
  This is the most important metric for road safety prediction.

- Fatal Precision: 0.04 — many false alarms when predicting Fatal. 
  The model over-predicts Fatal due to balancing.

- Logistic Regression struggles because severity prediction is a non-linear 
  problem. This result sets the baseline that Random Forest must beat.

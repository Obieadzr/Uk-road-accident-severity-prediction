```python
import pandas as pd 
import numpy as np
data = '../Data/cleaned_collision.csv'
df = pd.read_csv(data)
df.shape
```

```python
df['date']=pd.to_datetime(df['date'])
df['hour']=df['time'].str.split(':').str[0].astype(int)
df['month']=df['date'].dt.month
print(f"Hour new column \n{df['hour'].head()}")
print(f"Month new column \n{df['month'].head()}")
```

```python
df['season'] = df['month'].apply(lambda x : 0 if x in [12,1,2] else 1 if x in [3,4,5] else 2 if x in [6,7,8] else 3)
print(f"Season columns values \n{df['season'].value_counts().sort_index()}")
```

```python
df['is_rush_hour'] = df['hour'].apply(lambda x : 1 if x in [7,8,9,17,18,19] else 0)
print(f"{df['is_rush_hour'].value_counts().sort_index()}")
```

```python
df['day_name']=df['date'].dt.day_name()
print(df[['day_of_week','day_name']].value_counts().sort_index())
df['is_weekend'] = df['day_of_week'].apply(lambda x : 1 if x in [1,7] else 0)
print(df['is_weekend'].value_counts().sort_index())
df= df.drop('day_name',axis=1)
```

Why are we doing these step to see in day of week column we have number 1 to 7 to check what number represent what 

```python
print(df['light_conditions'].value_counts().sort_index())
df['is_dark'] = df['light_conditions'].apply(lambda x : 1 if x in [4,5,6,7] else 0)
df['is_dark'].value_counts()
```

```python
print(df['weather_conditions'].value_counts().sort_index())
df['is_bad_weather'] = df['weather_conditions'].apply(lambda x: 1 if x in [2,3,5,6,7] else 0)
print(df['is_bad_weather'].value_counts())
```

```python
print(df['speed_limit'].value_counts().sort_index())
df['is_highspeed'] = df['speed_limit'].apply(lambda x : 1 if x>=60 else 0)
df['is_highspeed'].value_counts()
```

```python
print(df['urban_or_rural_area'].value_counts())
df['is_urban'] = df['urban_or_rural_area'].apply(lambda x : 1 if x==1 else 0)
df['is_urban'].value_counts()
```

```python
print(df['junction_detail'].value_counts())
df['is_junction'] = df['junction_detail'].apply(lambda x : 0 if x == 0 else 1)
df['is_junction'].value_counts()
```

```python
print(df['carriageway_hazards'].value_counts().sort_index())
df['is_hazards'] = df['carriageway_hazards'].apply(lambda x : 0 if x==0 else 1)
df['is_hazards'].value_counts()
```

```python
df = df.drop(['date','time'],axis=1)
df.shape
```

```python
df.to_csv('../Data/enhanced_collision.csv',index= False)
print(df.shape)
```

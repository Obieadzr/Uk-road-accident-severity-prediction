#### Step 5 :- Scaling
###### Input :- enhanced_collision.csv
###### processing :- we are gonna scale 
###### Output :- We will get a new scaled dataset

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
df =pd.read_csv('../Data/enhanced_collision.csv')
df.shape
df.info()
```

```python
cols_toStandarize = ['longitude','latitude','speed_limit','number_of_vehicles','hour','month','number_of_casualties']
scale = StandardScaler()
for i in cols_toStandarize:
    df[i]=scale.fit_transform(df[i].values.reshape(-1,1))
print(df[cols_toStandarize].head())
```

```python
df.to_csv('../Data/Scaled_collision.csv',index=False)
print(df.shape)
```
